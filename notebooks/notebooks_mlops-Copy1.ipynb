```python
data1= pd.read_csv("result.csv", na_values=['?', None])
```

```python
data1=data1.set_index('auction_id')
data1.head()
```

## browser data

```python
dfbrowser= data1.drop(columns='browser')
data2 =dfbrowser
data2.head()
```

```python
# Let's create a Day of Week feature
data2['date']=pd.to_datetime(data2.date)
data2['day_of_week'] = data2['date'].dt.day_name()
# Let's drop date
del data2['date']
data2.head()
```

#### let's also drop columns yes and no which is  already replace by brand_Lux

```python

data2= data2.drop(['yes','no'],axis=1)
```

```python
data2.head()
```

```python
data2.info()
```

```python
#numerical variables
col_num= ['hour' ,'platform_os','brand_Lux']
#categorical variables
col_cat=['experiment','device_make','day_of_week']
```

```python
data2[col_cat]
```

```python
#one-hot encoding
d_cat=pd.get_dummies(data2[col_cat],drop_first=True)
d_cat
```

```python
data3 =pd.concat([d_cat,data2[col_num]],axis=1)
data3.columns.tolist
```

```python
data3.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling = MinMaxScaler()
data3[:] = scaling.fit_transform(data3[:])
data3.head()
```

```python
# Split data and split it to training set, validation set and testing set
import sklearn.utils
```

```python
data3 = sklearn.utils.shuffle(data3)
train, validate, test = np.split(data3.sample(frac=1), [int(.7*len(data3)), int(.9*len(data3))])
print('training set:', train.shape)
print('validation set:',validate.shape)
print('test set:', test.shape)
```

```python


X_train = train.iloc[:,0:123].values
Y_train = train.iloc[:,123].values

X_test= test.iloc[:,0:123].values
Y_test = test.iloc[:,123].values

X_val = validate.iloc[:,0:123].values
Y_val = validate.iloc[:,123].values
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
#Fitting Logistic Regression to the training set
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,Y_train)
```

```python
#Predicting the test set
y_pred=classifier.predict(X_test)
```

```python
y_pred
```

#### Let's make  the confusion matrix

```python
from sklearn.metrics  import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
```

### Decison Trees

```python
# Split dataset into training set and test set
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix


# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,Y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))
```
